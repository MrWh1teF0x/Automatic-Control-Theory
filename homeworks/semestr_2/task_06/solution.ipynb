## Задача 6.

Квадрокоптер выполняет задание удержания заданной высоты $h = 5 \, \text{м}$.  
Вертикальная динамика аппарата подвержена случайным порывам ветра, которые моделируются экспоненциально-коррелированным (Марковским) случайным процессом с постоянной корреляции $\tau_w = 2 \, \text{с}$.

Единственным датчиком является барометрический альтиметр, измерения которого содержат аддитивный белый гауссовский шум.

### Система уравнений:

$$
\begin{cases}
\dot{x} =
\begin{pmatrix}
0 & 1 & 0 \\
0 & 0 & 1 \\
0 & 0 & -0.5
\end{pmatrix}
x +
\begin{pmatrix}
0 \\
1 \\
0
\end{pmatrix}
u +
\begin{pmatrix}
0 \\
0 \\
1
\end{pmatrix}
v
\end{cases}
$$
### Вектор состояния системы:

$$
x(t) = [x_1, x_2, x_3]^T \in \mathbb{R}^3
$$

где:
- $x_1(t) = h(t) - h_{\text{зад}}(t)$ — отклонение текущей высоты от заданной $[м]$
- $x_2(t) = \dot{h}(t)$ — вертикальная скорость $[м/с]$
- $x_3(t) = a_w(t)$ — случайное ускорение от вертикального ветра $[м/с^2]$

### Параметры шумов:

**Шум объекта** $v(t)$:

$$
\mathbb{M}[v(t_1) v(t_2)] = V_w \delta(t_1 - t_2), \quad V_w = 0.04 \, \text{м}^2/\text{с}^4
$$

$$
\sigma_v = 0.2 \, \text{м}/\text{с}^2
$$

**Шум датчика** $w(t)$:

$$
\mathbb{M}[w(t_1) w(t_2)] = W_w \delta(t_1 - t_2), \quad W_w = 0.09 \, \text{м}^2
$$

$$
\sigma_w = 0.3 \, \text{м}
$$

## Постановка задачи:

Синтезировать оптимальный наблюдатель — фильтр Калмана.
