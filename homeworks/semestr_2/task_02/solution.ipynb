## Задача 2.

Матрицы системы продольного движения летательного аппарата, заданной в пространстве состояний, даны ниже:

$$
A = \begin{pmatrix} -0.01541 & -9.81 \\ 0.001712 & 0 \end{pmatrix}, \quad
B = \begin{pmatrix} -0.08 \\ 0.012 \end{pmatrix}, \quad
C = \begin{pmatrix} 1 & 0 \end{pmatrix}.
$$

## Постановка задачи:

1. Необходимо найти **регулятор Аккермана**, который будет располагать полюсы замкнутой системы в точке:

$$
\lambda_{1,2} = -0.25 \pm 0.20i
$$

т.е. найти вектор коэффициентов обратной связи по состоянию $K$ ($u = -Kx$).

2. Найти коэффициенты обратной связи для **наблюдателя Люенбергера** ($L$) для данной системы.

```python
import numpy as np
import pandas as pd

np.set_printoptions(
    precision=4, suppress=True, linewidth=100, threshold=1000, edgeitems=10
)

n = 2

A = np.array([[-0.01541, -9.81], [0.001712, 0]])

B = np.array([-0.08, 0.012])

C = np.array([[1, 0]])


# Матрица управляемости
M_c = np.array([np.linalg.matrix_power(A, i) @ B for i in range(n)]).T

rank_M_c = np.linalg.matrix_rank(M_c)

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.float_format", "{:.4f}".format)

df = pd.DataFrame(M_c, columns=[f"A^{i}*B" for i in range(n)])

print("Матрица управляемости:")
print()
print(df.to_string(index=False, header=True))

print()
print(f"Ранг матрицы управляемости (M_c) = {rank_M_c}")

print()
print()


# Матрица наблюдаемости
M_o = M_o = np.vstack([C @ np.linalg.matrix_power(A, i) for i in range(n)])

rank_M_o = np.linalg.matrix_rank(M_o)

df = pd.DataFrame(M_o)
df.index = [f"C*A^{i}" for i in range(n)]
df.columns = [""] * len(df.columns)

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.float_format", "{:.4f}".format)


print("Матрица наблюдаемости:")
print(df.to_string(header=True))

print()
print(f"Ранг матрицы наблюдаемости (M_o) = {rank_M_o}")

print()
print("=" * 50)

# Посчитаем вектор коэффициентов обратной связи для регулятора Аккермана

M_c_inv = np.linalg.inv(M_c)

Hi_A = np.linalg.matrix_power(A, 2) + 0.5 * A + 0.1025 * np.eye(2)

K = np.array([[0, 1]]) @ M_c_inv @ Hi_A

print()
print(f"K = {K[0]}")

# Посчитаем вектор коэффициентов обратной связи для наблюдателя Люенбергера

M_o_inv = np.linalg.inv(M_o)

L = Hi_A @ M_o_inv @ np.array([[0], [1]])

print()
print(f"L = {L.T[0]}")
```
