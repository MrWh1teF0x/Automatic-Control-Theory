## Задача 4.

Случайный процесс задан следующим образом:

$$
x(t) = A \cos(\omega_0 t) + B \sin(\omega_0 t)
$$

где $A$ и $B$ — независимые случайные величины со следующими свойствами:

$$
m_A = \mathbb{M}[A] = 2, \quad m_B = \mathbb{M}[B] = 0,
$$

$$
\sigma^2_A = D[A] = 4, \quad \sigma^2_B = D[B] = 4, \quad \mathbb{M}[AB] = 0,
$$

$$
\omega_0 = 3 \, \text{рад/с}.
$$

## Определить следующие характеристики случайного процесса:

1. Математическое ожидание  
2. Второй начальный вероятностный момент  
3. Дисперсию  
4. Корреляционную функцию  
5. Спектральную плотность
